```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
df=pd.read_csv('1_boston_housing.csv')
df
```

```python
X=df.loc[:,df.columns!='MEDV']
Y=df.loc[:,df.columns=='MEDV']

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
mms=MinMaxScaler()
mms.fit(X_train)
X_train=mms.transform(X_train)
X_test=mms.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(128,input_shape=(13,),activation='relu',name='dense1'))
model.add(Dense(64,activation='relu',name='dense2')) #dense1 and 2 activation=relu allows to learn complex pattern and introduce non-linaerity
model.add(Dense(1,activation='linear',name='output')) #activation=linear allows to make continuous prediction
model.compile(optimizer='adam',loss='mse')
model.summary()
```

```python
hs=model.fit(X_train,Y_train,epochs=100,validation_split=0.05,verbose=1)
```

```python
mse=model.evaluate(X_test,Y_test,verbose=0)
print(mse)
```
